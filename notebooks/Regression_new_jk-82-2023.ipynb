```python

def generate_stacked_reg(title, exps, dropcol=[], cov_order=[], order=False, siglevels=[0.05, 0.01, 0.001]):
    results=[]
    pseudo_r_list=[]
    column_names=[]
    for e in exps:
        
        if len(dropcol>0):
            drop_first=False
        else:
            drop_first=True
        print("running ", e['name'] )
        df_cat=pd.get_dummies(e['df'][e['cat']],columns=e['cat'], drop_first=drop_first)
        
        for col in dropcol:
            if col in df_cat.columns:
                df_cat.drop(col, axis=1, inplace=True)
        
        
        X=pd.concat([df_cat, e['df'][e['num']]], axis=1)
        #X=df_cat
        Xs = sm.add_constant(X)    # append constant(required for regression in python)
        column_names.append(e['name'])
        
        #check for missing values. Need to handle in preproccessing. 
        assert X.isna().sum().sum()==0
        
        #return X, e['df'][e['dv']]
        #return e['df'][e['dv']], X
        #return e['dv'], X_cons
        if e['type']=='bin':
            exp = sm.Logit(e['df'][e['dv']], Xs).fit()
            pseudo_r = exp.prsquared
            pseudo_r_list.append(round(pseudo_r, 4))
 
        elif e['type']=='ols':
            exp = sm.OLS(e['df'][e['dv']], Xs).fit()
        results.append(exp)
        column_names.append(e['name'])
        
   
    
    stargazer = Stargazer(results)
    #stargazer.custom_columns(column_names,[1 for x in range(len(column_names))])
    stargazer.significance_levels([0.05, 0.01, 0.001])
    if order:
        stargazer.covariate_order(cov_order)
    return results, stargazer, X.columns

cat_controls= ['vehicle_vin', #supply
               'auction_is_green_light',   #condition
               'auction_is_yellow_light',  #condition
               'auction_is_blue_light',    #condition
               'auction_red_light_changed_to_green_in_past',
              'week_num','auction_hour_of_week'] #condition] 

        #        df_cat.drop(col, axis=1, inplace=True)

               
num_controls = ['auction_gross_run_number', 
                'auction_odometer'] 

dvs=['auction_floor_price','auction_is_gross_sold','auction_sale_amount']  

all_var=cat_controls+num_controls+dvs
df=df[all_var]
ex1a={'name': 'Floor Price','df':df,'cat': cat_controls, 'num':num_controls, 'dv': 'auction_floor_price', 'type':'ols'}
ex1b={'name': 'Logistic','df':df,'cat': cat_controls, 'num':num_controls, 'dv': 'auction_is_gross_sold', 'type':'bin'}
ex1c={'name': 'Price','df':df[df['auction_is_gross_sold']==1],'cat': cat_controls, 'num':num_controls, 'dv': 'auction_sale_amount', 'type':'ols'}

exps=[ex1a,ex1c]  #ex1b isn't working Not sure why yet

               
results1, table1, columns1 =generate_stacked_reg('All', exps, dropcol=['auction_hour_of_week_0', 'vehicle_vin_00000']   order=False,)
table1
```
